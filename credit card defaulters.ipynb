```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,8),dpi=100)
sns.heatmap(df.corr(),annot=True)
```

```python
df['bill_amt1']=df['bill_amt1'].astype(int)
df['bill_amt2']=df['bill_amt2'].astype(int)
df['bill_amt3']=df['bill_amt3'].astype(int)
df['bill_amt4']=df['bill_amt4'].astype(int)
df['bill_amt5']=df['bill_amt5'].astype(int)
df['bill_amt6']=df['bill_amt6'].astype(int)
df['pay_amt1']=df['pay_amt1'].astype(int)
df['pay_amt2']=df['pay_amt2'].astype(int)
df['pay_amt3']=df['pay_amt3'].astype(int)
df['pay_amt4']=df['pay_amt4'].astype(int)
df['pay_amt5']=df['pay_amt5'].astype(int)
df['pay_amt6']=df['pay_amt6'].astype(int)
```

```python
total_bill_amt=df.loc[:,['bill_amt1','bill_amt2','bill_amt3','bill_amt4','bill_amt5','bill_amt6']].sum(axis=1)
```

```python
df['total_bill_amt']=total_bill_amt
```

```python
total_pay_amt=df.loc[:,['pay_amt1','pay_amt2','pay_amt3','pay_amt4','pay_amt5','pay_amt6']].sum(axis=1)
```

```python
df['total_pay_amt']=total_pay_amt
```

```python
df['default_amt']=df['total_bill_amt']-df['total_pay_amt']
```

```python
df
```

```python
df['default.payment.next.month'].corr(df['default_amt'])
```

```python
features=df.drop('default.payment.next.month',axis=1)
```

```python
target=df['default.payment.next.month']
```

```python
from sklearn.feature_selection import mutual_info_classif
imp=mutual_info_classif(features,target)
```

```python
feature_importance=pd.Series(imp,df.columns[0:len(df.columns)-1])
```

```python
feature_importance
```

```python
feature_importance.plot(kind='barh')
```

```python
# age and sex has very less importance
```

```python
df.drop(['age','sex'],axis=1,inplace=True)
```

```python
sns.countplot(hue=df['default.payment.next.month'],x=df['education'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size = 0.2)
```

```python
from sklearn.metrics import classification_report,f1_score
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression() 

lr_model.fit(x_train,y_train)

y_pred = lr_model.predict(x_test) 

print(f"F1 score:{f1_score(y_test,y_pred)}")
print(f"Classification Report for logistic regression:{classification_report(y_test,y_pred)}")
```

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(x_train,y_train)

y_pred_tree = tree_model.predict(x_test)

print(f"F1 score:{f1_score(y_test,y_pred_tree)}")
print(f"Report :{classification_report(y_test,y_pred_tree)}")
```
